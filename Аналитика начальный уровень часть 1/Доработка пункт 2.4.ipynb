```python
# Подготовка данных
import pandas as pd

df_courses = pd.read_csv("./Data/courses.csv", encoding="UTF-8")
df_course_contents = pd.read_csv("./Data/course_contents.csv", encoding="UTF-8")
df_progresses = pd.read_csv("./Data/progresses.csv", encoding="UTF-8")
df_progress_phases = pd.read_csv("./Data/progress_phases.csv", encoding="UTF-8")

result = pd.merge(df_progress_phases, df_progresses, how='left', left_on='progress_id', right_on='id')
result2 = pd.merge(result, df_courses, how='left', left_on=['course_id'], right_on=['id'])
main = pd.merge(result2, df_course_contents, how='left', 
                   left_on=['course_id', 'module_number', 'lesson_number'], 
                   right_on=['course_id', 'module_number', 'lesson_number'])
del main['id_y']
del main['id_x']
del main['Unnamed: 0']
main.head()
```

```python
# функция пригодится позже
from datetime import datetime, MAXYEAR

def transformation(x):
    try:
        if x != None:
            x = datetime.strptime(str(x[:x.find(' ')]),"%Y-%m-%d").strftime("%Y%m") 
        else:
#             x = int(str(MAXYEAR) +'12')
            x = int(201907)
        
    except AttributeError:
#         x = int(str(MAXYEAR) +'12')
        x = int(201907)
    return x
```

```python
# выберем нужные столбцы
df = main.loc[:, ['status', 'title', 'start_date', 'finish_date', 'progress_id', 'is_homework']].sort_values('start_date', axis = 0)

# выбираю данные с домашней работой
df = df[df['is_homework'] == True]

df['start_date_month'] = df.loc[:, 'start_date'].apply(transformation)
df['finish_date_month'] = df.loc[:, 'finish_date'].apply(transformation)

# для удобного рассчета приведу столбцы к типу int
df['start_date_month'] = df['start_date_month'].astype(int)
df['finish_date_month'] = df['finish_date_month'].astype(int)

# уберу не нужные колонки
del df['start_date']
del df['finish_date']
del df['is_homework']
```

```python
# сгруппирую строки, что бы за одну дату и курс не было несколько одинаковых прогрессов
# подсчитаем количество прогрессов, домашки начатые и не оконченные в одном месяце пойдут в общие
# общие будут под максимальной датой
df_identical_month_gr = df.groupby(['course_id', 'title', 'start_date_month', 'finish_date_month', 'progress_id']
                                  ).count().reset_index().sort_values(['start_date_month','course_id'])

# Теперь когда у нас нет дублей по прогрессам, по периоду и курсу, то можно их посчитать
# Делаю группировку по нужным мне колонкам
df_identical = df_identical_month_gr.loc[:, ['course_id', 'title', 'finish_date_month', 'progress_id']].groupby(
    ['course_id', 'title', 'finish_date_month']
    ).count().sort_values(['finish_date_month', 'title']).reset_index()

# выведу первые 20 строк с количеством прогрессов по выполнению домашних работ в каждом месяце за всю историю
df_identical.head(20)
```
